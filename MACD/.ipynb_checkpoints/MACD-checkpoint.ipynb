```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', ''))
    return(x)
```

```python
# df = pd.read_csv("bitcoin.csv")
# df_price = df['Weighted_Price'].dropna()
df = pd.read_csv("apple.csv")
df[' Close/Last'] = df[' Close/Last'].apply(clean_currency).astype('float')
df_price = df[' Close/Last'].dropna()
price_list = df_price.tolist()
price_list.reverse()
df_price_list = pd.DataFrame(price_list)
df_price_list
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df_price_list)
plt.show()
```

```python
def ema(s, n):
    """
    returns an n period exponential moving average for the time series s

    s is a list ordered from oldest (index 0) to most recent (index -1)
    n is an integer

    returns a numeric list of the exponential moving average
    """
    ema = []
    j = 1

    #get n sma first and calculate the next n period ema
    sma = sum(s[:n])/n
    multiplier = 2/float(1+n)
    ema.append(sma)

    #EMA(current) = ( (Price(current) - EMA(prev) ) x Multiplier) + EMA(prev)
    ema.append(((s[n]-sma)*multiplier)+sma)

    #now calculate the rest of the values
    for i in s[n+1:]:
        tmp = ((i-ema[j])*multiplier)+ema[j]
        j = j + 1
        ema.append(tmp)
    return ema
```

```python
period = 1200

ema12 = ema(price_list, 12)
ema26 = ema(price_list, 26)
ema12_period = ema12[-period:]
ema26_period = ema26[-period:]
price_list_period = price_list[-period:]
macd = [ema12_i - ema26_i for ema12_i, ema26_i in zip(ema12_period, ema26_period)]
macd9 = ema(macd, 9)
macd = macd[9:]
macd_hist = [macd_i - macd9_i for macd_i, macd9_i in zip(macd, macd9)]
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize = (20, 10))

ax1.plot(ema12_period, label='ema12')
ax1.plot(ema26_period, label='ema26')
ax1.plot(price_list_period, label='price')
ax1.legend()

ax2.plot(macd, label='macd')
ax2.plot(macd9, label='signal line')
x, y = [0, len(macd)], [0, 0]
ax2.plot(x, y)
x_range = np.arange(0, len(macd), 1)
ax2.bar(x_range, macd_hist, color='red', alpha=0.5, label='macd histogram')
ax2.legend()

plt.show()
```

```python
def sum_list(list):
    result = 0
    for i in range(0, len(list)):
        result += list[i]
    return result
price_exchange = price_list_period[9:]
print(len(price_exchange))
print(len(macd_hist))
```
