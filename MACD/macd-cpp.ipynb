```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot data generated by MACD c++ version
df_cpp = pd.read_csv('macd.csv', names=['Price'], header=None)
macd_cpp = df_cpp.Price.tolist()
df_decisions = pd.read_csv('decisions.csv', names=['Decision'], header=None)
decisions = df_decisions.Decision.tolist()
df_tanh = pd.read_csv('tanh_decisions.csv', names=['Decision'], header=None)
tanh = df_tanh.Decision.tolist()
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize = (20, 10))
x_range = np.arange(0, len(macd_cpp), 1)
x, y = [0, len(macd_cpp)], [0, 0]

ax1.plot(x, y)
ax1.bar(x_range, macd_cpp, color='red', alpha=0.5, label='macd')
ax1.bar(x_range, decisions, color='green', alpha=0.5, label='decisions')
ax1.set_title('Desicions')
ax1.legend()

ax2.bar(x_range, tanh, color='purple', alpha=0.5, label='tanh')
ax2.bar(x_range, macd_cpp, color='red', alpha=0.5, label='macd')
ax2.set_title('tanh decisions')
ax2.legend()

plt.show()
```

```python
errors = [];
error_sum = 0;
for i in range(0, len(decisions)):
    if decisions[i] != tanh[i]:
        errors.append(tanh[i])
        error_sum += 1;
    else:
        errors.append(0);
print('error: ', error_sum);
```

```python
plt.figure(figsize = (20, 10))
x_range = np.arange(0, len(macd_cpp), 1)
x, y = [0, len(macd_cpp)], [0, 0]

plt.plot(x, y)
plt.bar(x_range, macd_cpp, color='red', alpha=0.5, label='macd')
plt.bar(x_range, errors, color='green', alpha=0.5, label='errors')
plt.legend()
plt.show()
```
